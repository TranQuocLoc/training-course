## Assignment Operators (Gán biến)

When assigning something to a variable using a ```=```, the right side of the ```=``` is always executed before the left side. This means that we can do something with a variable on the right side, then assign the result back to the same variable on the left side.

````python
a = 1
b = a
a = a + 1
c = a
print(a,b,c)
````

# CONDITIONAL (IF, ELSE, ELIF)

The logical conditions ```==, !=, <, >, >=, <=``` can be used in several ways, most commonly in "if statements" and loops. <br>
An "if statement" is written by using the ```if``` keyword.

````python
a = 33
b = 200
if b > a:
  print("b is greater than a")
````

The ```elif``` keyword is Python's way of saying "if the previous conditions were not true, then try this condition".

````python
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
````

The ```else``` keyword catches anything which isn't caught by the preceding conditions.

````python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
````

````python
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("a is greater than or equal to b")
````

# LOOPS (VÒNG LẶP)

## While

With the ```while``` loop we can execute a set of statements as long as a condition is true.

````python
i = 1
while i < 6:
  print(i)  # Print i as long as i is less than 6:
  i += 1  # increment i
````

```break```: stop the loop even if the while condition is true

````python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
````

```continue``` stop the current iteration, and continue with the next

````python
i = 1
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
````

## For

A ```for``` loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

````python
# Looping through a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)
````

````python
# Looping Through a String
for x in "banana":
  print(x)
````

```range()```

````python
for x in range(7): # x chạy từ 0 -> 6-1=5
  print(x)
````

````python
for number in range(2, 6): # x chạy từ 2 đến 5
  print(number)
````

````python
for x in range(30, 2, -3): # 2: index đầu, 30: index cuối, 3: step
  print(x)
````
