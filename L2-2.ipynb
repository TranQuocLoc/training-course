````python
df = pd.read_csv('Iris.csv')
df["Species"].value_counts()
````

````python
x1 = df.loc[df['Species']=='Iris-setosa']['PetalLengthCm']
y1 = df.loc[df['Species']=='Iris-setosa']['PetalWidthCm']
x2 = df.loc[df['Species']=='Iris-versicolor']['PetalLengthCm']
y2 = df.loc[df['Species']=='Iris-versicolor']['PetalWidthCm']
x3 = df.loc[df['Species']=='Iris-virginica']['PetalLengthCm']
y3 = df.loc[df['Species']=='Iris-virginica']['PetalWidthCm']
````
