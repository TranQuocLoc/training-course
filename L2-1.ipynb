# Cấu trúc dữ liệu (DATA STRUCTURE)

Cấu trúc dữ liệu là một cách tổ chức dữ liệu để có thể truy cập dữ liệu hiệu quả hơn tùy thuộc vào tình huống

## Danh sách (list)

Danh sách (list) là tập hợp dữ liệu có thứ tự. Các phần tử trong danh sách không cần phải cùng loại.<br>
Tính chất: **ordered** and **changeable**. **Allows duplicate members**.

### Tạo list

````python
# Tạo list bằng dấu ngoặc vuông []
my_list = [1, 2, 3 ,"linh", True]
print(my_list)
list2 = [1,2,3,1,1,1]
print(list2)
# Tạo list bằng lệnh list và ngoặc tròn (())
your_list = list((1, 2, 3 ,"linh", True))
print(your_list)
````

````python
print(type(my_list))
````

````python
# Độ dài của list: len()
print(len(my_list))
````

### Truy cập phần tử của list

![list](list.png)

````python
# Truy cập MỘT phần tử của list

print("Truy cập phần tử đầu tiên: ")
print(my_list[0])
print("Truy cập phần tử thứ hai: ")
print(my_list[1])
print("Truy cập phần tử cuối cùng: ")
print(my_list[-1])
print("Truy cập phần tử cuối cùng thứ hai: ")
print(my_list[-2])
````

````python
# Truy cập NHIỀU phần tử của list
print("Truy cập phần tử thứ hai, ba, bốn của list: ")
print(my_list[1:4])   # Bắt đầu từ index 1 (bao gồm) và kết thúc ở index 4 (không bao gồm)

print("Truy cập từ phần tử đầu tiên đến một index: ")
print(my_list[:4])   # Lấy ra phần từ thứ 1, 2, 3

print("Truy cập từ một index đến phần tử cuối cùng: ")
print(my_list[2:])
````

### Biến đổi list

#### Thay đổi phần tử

````python
# Thay đổi một phần tử trong list
shop_list = ['orange', 'apple', 'meat', 'tomato']
shop_list[2] = 'pork'
print(shop_list)
````

````python
# Thay đổi nhiều phần tử trong list
shop_list[1:3] = ['cucumber', 'fish']
print(shop_list)
````

#### Thêm phần tử

````python
# Thêm MỘT phần tử vào CUỐI list, ta dùng append()
shop_list.append('rice')
print(shop_list)
````

````python
# Thêm các phần tử từ list khác vào CUỐI một list, ta dùng extent()
sport_list = ['volleyball', 'badminton']
shop_list.extend(sport_list)
print(shop_list)
# shop_list.append(sport_list)
# print(shop_list)
````

#### Xóa phần tử

````python
# Xóa một phần tử cụ thể, dùng remove(). Nếu có nhiều hơn 1 phần tử, xóa phần tử đầu tiên tìm thấy
my_list = ['bach', "khang", "linh", 'hoa', 'linh']
my_list.remove('linh')
print(my_list)
````

````python
# Xóa một index cụ thể, dùng pop()
my_list.pop(1)
print(my_list)
````

Condition

[*expression* for *item* in *iterable* if *condition* == True]: accepts the items that evaluate to "True".

````python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if x != "apple"]
print(newlist)

newlist = []
for x in range(len(fruits)):
    if x != "apple":
        newlist.append(fruits[x])
````

````python
newlist = [x for x in range(10) if x < 5]
````

````python
newlist = ["Hello" for x in range(10) if x < 5]
print(newlist)
````

## Dictionary

Từ điển được sử dụng để lưu trữ giá trị dữ liệu theo cặp "khóa:giá trị". <br>
Tính chất: ordered, changeable and do not allow duplicates.

### Tạo dict

````python
thisdict = {
  "brand": "Ford",     # key : value
  "model": "Mustang",   # key : value
  "year": 1964          # key : value
}
print(thisdict)
````

item = key : value

````python
# Chú ý không được lặp key
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
````

````python
# Các value có thể thuộc bất cứ data type nào
thisdict = {
  "brand": "Ford",  
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
````

#### Get keys

````python
x = thisdict.keys()
print(x)
````

#### Get values


````python
x = thisdict.values()
print(x)
````

````python
# Lấy value của một key
print(thisdict["brand"])
# hoặc 
x = thisdict.get("model")
print(x)
````

#### Get items

````python
x = thisdict.items()
print(x)
````

### Change / add / remove items

````python
student = {
    "name": "Linh",
    "sex": "Female",
    "address": "Hanoi",
    "age": 23
}
# Change value
student["age"] = 24
# update()
student.update({"name": "Hoa"})
print(student)
````

````python
# Add items
student["graduate"] = True
print(student)
# Remove items, dùng pop()
student.pop("sex")
print(student)
````

### Nested dictionary

````python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
````

````python
# Access items
print(myfamily["child2"]["year"])
````

## Một số hàm thống kê

```arr.max()``` hoặc ```np.max(arr)```: Lấy giá trị lớn nhất của mảng arr.<br>
```arr.min()``` hoặc ```np.min(arr)```: Lấy giá trị nhỏ nhất của mảng arr.<br>
```arr.sum()``` hoặc ```np.sum(arr)```: Tổng tất cả các phần tử trong mảng arr.<br>
```arr.mean()``` hoặc ```np.mean(arr)```: Trung bình cộng của tất cả các phần tử trong mảng arr.<br>
```np.median(arr)```: Trả về giá trị trung vị của mảng arr.

````python
arr = np.array([1, 2, 3, 4, 5])
print("Giá trị lớn nhất của mảng arr là:", np.max(arr))
print("Giá trị nhỏ nhất của mảng arr là:", np.min(arr))
print("Tổng tất cả các phần tử của mảng arr là:", np.sum(arr))
print("Trung bình cộng tất cả các phần tử của mảng arr là:", np.mean(arr))
print("Giá trị trung vị của mảng arr là:", np.median(arr))
````
